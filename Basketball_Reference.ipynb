Scraping Table from a website



```python
import requests
from bs4 import BeautifulSoup
from string import ascii_lowercase
```

```python
player_saved_data = ""

for letter in ascii_lowercase:

  source = requests.get(f'https://www.basketball-reference.com/players/{letter}/').text

  soup = BeautifulSoup(source, 'lxml')

  for record in soup.find_all('tr'):

    player_data = ""

    for x in record.find_all('th', scope = 'row'):

      player_data = x.find('a').text

    for data in record.find_all('td'):
      player_data = player_data + ',' + data.text

    player_saved_data = player_saved_data + '\n' + player_data
  
print(player_saved_data)
```

```python
header = "Player,From,To,Pos,Ht,Wt,Birth Date,Birth Year,Colleges"
```

```python
with open('Basketball.txt', 'w', encoding='ascii', errors='ignore') as r:
    r.write(header)
    r.write(player_saved_data)
```
